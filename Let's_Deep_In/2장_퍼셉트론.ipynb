# 2강 - 퍼셉트론
- 퍼셉트론은 다수의 신호를 받아서 하나의 신호를 출력하는 것을 말한다. 딥러닝의 기원이 되는 알고리즘임.
- 퍼셉트론에서는
가중치와
편향을 매개변수로 설정
- 퍼셉트론으로
AND, OR
게이트 등의 논리 회로를
표현 할 수 있음
- 단층 퍼셉트론으로는 XOR게이트를 구현할 수 없다. 그래서 다층 퍼셉트론의 개념이 등장했다.
- 단층
퍼셉트론은 직선형
영역만 표현할 수 있고 ,
다층 퍼셉트론은 비선형
영역도 표현할 수 있음

## 단순 논리 회로
1. AND
2. OR
3. NAND

```
import numpy as np
```

```
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
```

```
def AND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```
print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```

```
def NAND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5])
  b = 0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```
print(NAND(0, 0))
print(NAND(1, 0))
print(NAND(0, 1))
print(NAND(1, 1))
```

```
def OR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.2
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```
print(OR(0, 0))
print(OR(1, 0))
print(OR(0, 1))
print(OR(1, 1))
```

## 다층 퍼셉트론의 구현
- 기존의
단층 퍼셉트론으로는
XOR
게이트를 표현 할 수 없다
- 다층
퍼셉트론으로 XOR 을
구현
해 볼 수 있다

image.png

```
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```

```
print(XOR(0, 0))
print(XOR(1, 0))
print(XOR(0, 1))
print(XOR(1, 1))
```

## XNOR 구현해보기
- XNOR구현을 위해서는 NOR이 필요하다 NOR은 OR의 not으로 x1, x2 값이 0, 0 일때에만 1을 출력한다.
image.png

```
def NOR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5])
  b = 0.2
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```
print(NOR(0, 0))
print(NOR(1, 0))
print(NOR(0, 1))
print(NOR(1, 1))
```

```
def XNOR(x1, x2):
  s1 = NOR(x1, x2)
  s2 = AND(x1, x2)
  y = OR(s1, s2)
  return y
```

```
print(XNOR(0, 0))
print(XNOR(1, 0))
print(XNOR(0, 1))
print(XNOR(1, 1))
```
