# 1.활성화 함수
- 임계값을 경계로 출력이 바뀌는데 이런 함수를 **계단함수** 라고 한다. (0과 1중 하나의 값만 돌려준다.)
- 시그모이드 함수는 부드러운 곡선이며 입력에 따라 출력이 연속적으로 변화한다.
- 둘다 비선형 함수이고, 입력값이 작으면 0에 가깝게 출력(혹은 0) 입력값이 커지면 출력 1에 가까워지는 (혹은 1) 구조.

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def step_function(x):
  if x>0:
    return 1
  else:
    return 0
```

```python
def step_function(x):
  y = x>0 #True or False 값을 가진다.
  return y.astype(np.int) # True면 1 False면 0
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show
```

## Sigmoid
image.png

```python
def sigmoid(x):
  return 1 / (1+np.exp(-x))
```

```python
x = np.array([-1.0, 1.0, 2.0])
sigmoid(x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show
```

## Relu
image.png

```python
def relu(x):
  return np.maximum(0, x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.ylim(-0.1, 5.1)
plt.show
```

## tanh
image.png

```python
def tanh(x):
  return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = tanh(x)
plt.plot(x, y)
plt.ylim(-1.1, 1.1)
plt.show
```
