#문제3

```
df4= df.loc['서울':'제주','2012-12-01':]
df4= df4.div(df4['2012-12-01'],axis=0)
df4
```

```
f,ax = plt.subplots(2,1,figsize=(10,16))
df4.T.plot(ax=ax[0])
df4['2020-09-01'].sort_values(ascending=False).plot.bar(ax=ax[1])
ax[1].set_ylim([0,4.0])
ax[1].set_title('2012년도 대비 2020년 지역별 가격변동')
plt.show()
```

#문제5

```
idx = df.index
df['Dist'] = df.index.str[:2]
df5 = df.loc[:,'2020-09-01':]
categories=df5['Dist'].unique()
df5
```

```
df6=[]
```

```
for i in categories:
  df6.append(df5[df['Dist']==i]['2020-09-01'].max()-df5[df['Dist']==i]['2020-09-01'].min())
```

```
data= pd.DataFrame(data={"Max-min":df6},index=categories)
```

```
data
```

```
f,ax = plt.subplots(1,1,figsize=(10,8))
data.sort_values(by='Max-min',ascending=False).plot.bar(ax=ax)
plt.show()
```
